```python
import cv2
import seaborn as sns
import os, glob, shutil
import matplotlib.pyplot as plt
from skimage.feature import greycomatrix, greycoprops
import numpy as np
```

```python
img_path = r"D:\final_year_project\dataset_900\DiPASL-S900\N\Image_1687588651.8768187.png"
img0 = cv2.imread(img_path)
img = cv2.cvtColor(img0, cv2.COLOR_RGB2BGR)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

# Texture Analysis using GLCM (Grey-Level Co-occurrence Matrix)
# Calculate GLCM properties
distances = [1]  # Distance between pixels
angles = [0, np.pi/4, np.pi/2, 3*np.pi/4]  # Directions of pixel pairs
glcm = greycomatrix(img, distances, angles, levels=256, symmetric=True, normed=True)
contrast = greycoprops(glcm, 'contrast')
homogeneity = greycoprops(glcm, 'homogeneity')

# Print the histogram and texture analysis results
print("Histogram:")
print(hist)

print("\nTexture Analysis - Contrast:")
print(contrast)

print("\nTexture Analysis - Homogeneity:")
print(homogeneity)
```

```python
image = img
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Histogram Analysis
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])

# Texture Analysis using GLCM (Grey-Level Co-occurrence Matrix)
# Calculate GLCM properties
distances = [1]  # Distance between pixels
angles = [0, np.pi/4, np.pi/2, 3*np.pi/4]  # Directions of pixel pairs

glcm_r = greycomatrix(gray[:, :, 0], distances, angles, levels=256, symmetric=True, normed=True)
glcm_g = greycomatrix(gray[:, :, 1], distances, angles, levels=256, symmetric=True, normed=True)
glcm_b = greycomatrix(gray[:, :, 2], distances, angles, levels=256, symmetric=True, normed=True)

contrast_r = greycoprops(glcm_r, 'contrast')
contrast_g = greycoprops(glcm_g, 'contrast')
contrast_b = greycoprops(glcm_b, 'contrast')

homogeneity_r = greycoprops(glcm_r, 'homogeneity')
homogeneity_g = greycoprops(glcm_g, 'homogeneity')
homogeneity_b = greycoprops(glcm_b, 'homogeneity')

# Print the histogram and texture analysis results
print("Histogram:")
print(hist)

print("\nTexture Analysis - Contrast:")
print("Red:", contrast_r)
print("Green:", contrast_g)
print("Blue:", contrast_b)

print("\nTexture Analysis - Homogeneity:")
print("Red:", homogeneity_r)
print("Green:", homogeneity_g)
print("Blue:", homogeneity_b)
```
